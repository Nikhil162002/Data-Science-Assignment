```python
from sklearn.preprocessing import MultiLabelBinarizer

   # Assuming 'genre' is the column name containing genres as strings
mlb = MultiLabelBinarizer()
genre_encoded = mlb.fit_transform(df['genre'].str.split(', '))  # Split genres by comma and space

   # Create a new DataFrame with encoded genres
genre_df = pd.DataFrame(genre_encoded, columns=mlb.classes_, index=df.index)

   # Concatenate the encoded genres with the original DataFrame
df = pd.concat([df, genre_df], axis=1)
```

```python
df = pd.get_dummies(df, columns=['type', 'genre'], prefix=['type', 'genre'])  # Adjust column names as needed
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['rating_normalized'] = scaler.fit_transform(df[['rating']])
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def recommend_anime(target_anime_title, df, top_n=10, threshold=0.8):
    """
    Recommends anime based on cosine similarity.

    Args:
        target_anime_title: The title of the anime to find recommendations for.
        df: The DataFrame containing the anime data with features.
        top_n: The maximum number of recommendations to return.
        threshold: The minimum cosine similarity score for an anime to be considered similar.

    Returns:
        A list of recommended anime titles.
    """

    # Get the feature vector for the target anime
    target_anime = df[df['name'] == target_anime_title]  # Assuming 'name' is the column for anime titles
    if target_anime.empty:
        return "Anime not found in the dataset."  # Handle case where target anime isn't in the dataset

    target_features = target_anime.drop(columns=['name',  'episodes', 'rating', 'members' ])  # Drop non-feature columns

    # Calculate cosine similarity with all other anime
    similarity_scores = cosine_similarity(target_features, df.drop(columns=['name',  'episodes', 'rating', 'members']))[0]

    # Get indices of similar anime (excluding the target anime itself)
    similar_anime_indices = np.argsort(similarity_scores)[::-1][1:]  # Sort and exclude the target anime

    # Filter recommendations based on the threshold
    filtered_indices = [i for i in similar_anime_indices if similarity_scores[i] >= threshold]

    # Get the titles of the recommended anime
    recommended_anime = df.iloc[filtered_indices]['name'].head(top_n).tolist()

    return recommended_anime
```

```python
# Example usage
recommendations = recommend_anime('Death Note', df, top_n=5, threshold=0.7)
print(recommendations)  # To see the output, run the code.
```
