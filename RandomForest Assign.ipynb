```python
def outlier_detection(data,column):
  q1 = data[column].quantile(0.25)
  q3 = data[column].quantile(0.75)
  iqr = q3-q1

  lower_ext= q1-iqr*1.5
  upper_ext = q3+iqr*1.5
  return lower_ext,upper_ext
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_col = ['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe']
```

```python
scaler = StandardScaler()

df[numerical_col] = scaler.fit_transform(df[numerical_col])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Type', axis=1)
y=df['Type']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier= RandomForestClassifier(n_estimators = 100, random_state = 0)


rf_classifier.fit(x_train,y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
precision = precision_score(y_test,y_pred, average = 'micro')
print(precision)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=0)
rf_classifier.fit(x_train, y_train)
y_pred_rf = rf_classifier.predict(x_test)
```

```python
# Bagging Classifier (using Decision Tree as base estimator)
bagging_classifier = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, random_state=0)
bagging_classifier.fit(x_train, y_train)
y_pred_bagging = bagging_classifier.predict(x_test)
```

```python
result = cross_val_score(bagging_classifier,x,y,cv=KFold(n_splits=10))
```

Additional Notes

Boosting : Boosting involves training base estimators sequentially, where each subsequent estimator focuses on correcting the errors made by the previous ones.

Bagging : Bagging involves creating multiple subsets of the original dataset by randomly sampling data points with replacement (bootstrapping).
