```python
def outlier_detection(data,column):
  q1 = data[column].quantile(0.25)
  q3 = data[column].quantile(0.75)
  iqr = q3- q1
  upper_ext = q3 + iqr*1.5
  lower_ext = q1 - iqr*1.5
  return lower_ext,upper_ext
```

```python
outlier_detection(df,'Age')
```

**Building and Evaluating LightGBM and XGBoost Models**

```python
from sklearn.model_selection import train_test_split

x = df.drop('Survived',axis =1)
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
import lightgbm as lgb
import xgboost as xgb

# Create LightGBM model
lgb_model = lgb.LGBMClassifier(random_state=0)

# Create XGBoost model
xgb_model = xgb.XGBClassifier(random_state=0)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Train LightGBM model
lgb_model.fit(x_train, y_train)
lgb_predictions = lgb_model.predict(x_test)

# Evaluate LightGBM model
lgb_accuracy = accuracy_score(y_test, lgb_predictions)
# ... (calculate other metrics similarly)

# Train XGBoost model
xgb_model.fit(x_train, y_train)
xgb_predictions = xgb_model.predict(x_test)

# Evaluate XGBoost model
xgb_accuracy = accuracy_score(y_test, xgb_predictions)
# ... (calculate other metrics similarly)

# Print the evaluation results
print("LightGBM Accuracy:", lgb_accuracy)
# ... (print other metrics)
print("XGBoost Accuracy:", xgb_accuracy)
# ... (print other metrics)
```

**Optimizing Models with Cross-Validation and Hyperparameter Tuning**

```python
from sklearn.model_selection import cross_val_score
```

```python


# Perform 5-fold cross-validation for LightGBM
lgb_scores = cross_val_score(lgb_model, x, y, cv=5, scoring='accuracy')  # Use desired scoring metric
print("LightGBM Cross-Validation Scores:", lgb_scores)
print("LightGBM Mean Accuracy:", lgb_scores.mean())

# Perform 5-fold cross-validation for XGBoost
xgb_scores = cross_val_score(xgb_model, x, y, cv=5, scoring='accuracy')  # Use desired scoring metric
print("XGBoost Cross-Validation Scores:", xgb_scores)
print("XGBoost Mean Accuracy:", xgb_scores.mean())
```

Hyperparameter Tuning (using GridSearchCV):

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grids for LightGBM and XGBoost
lgb_param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.05, 0.01],
    # ... other LightGBM hyperparameters
}

xgb_param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.1, 0.05, 0.01],
    # ... other XGBoost hyperparameters
}

# Create GridSearchCV objects
lgb_grid = GridSearchCV(lgb_model, lgb_param_grid, cv=5, scoring='accuracy')  # Use desired scoring metric
xgb_grid = GridSearchCV(xgb_model, xgb_param_grid, cv=5, scoring='accuracy')  # Use desired scoring metric

# Fit the GridSearchCV objects to find the best hyperparameters
lgb_grid.fit(x_train, y_train)
xgb_grid.fit(x_train, y_train)

# Get the best models and their scores
best_lgb_model = lgb_grid.best_estimator_
best_xgb_model = xgb_grid.best_estimator_

print("Best LightGBM Parameters:", lgb_grid.best_params_)
print("Best LightGBM Score:", lgb_grid.best_score_)
print("Best XGBoost Parameters:", xgb_grid.best_params_)
print("Best XGBoost Score:", xgb_grid.best_score_)
```

By combining cross-validation and hyperparameter tuning, you can significantly improve the performance and generalization ability of your LightGBM and XGBoost models.
