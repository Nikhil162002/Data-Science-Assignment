```python
data = [1.13,1.55,1.43,0.92,1.25,1.36,1.32,0.85,1.07,1.48,1.20,1.33,1.18,1.22,1.29]
print(data)
```

```python
import scipy.stats as stats

sample_mean = np.mean(data)
sample_std = np.std(data,ddof = 1)

print(sample_mean,sample_std)
```

```python
# Calculate the critical value for a 99% confidence interval
confidence_level = 0.99
alpha = 1 - confidence_level
degrees_of_freedom = len(data) - 1
t_stat = stats.t.ppf(1-alpha/2,degrees_of_freedom)
t_stat
```

```python
margin_of_error = t_stat*(sample_std/np.sqrt(len(data)))
margin_of_error
```

```python
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(confidence_interval)
```

```python
# exercise 1: Prime Numbers

def prime(number):
  """check if a number is prime."""

  if number<=1:
    return False
  for i in range(2, int(number**0.5) +1):
    if number % i ==0:
      return False
    return True

number = int(input('Enter a number:'))
if prime(number):
  print(' it is a prime number')
else:
  print("it is not a prime number")
```

```python
# exercise 4: Word counter
def count_words(text):
  words = text.split()
  word_counts = {}

  for word in words:
    word_counts[word] = word_counts.get(word, 0) +1

  return word_counts

input_text = 'this is a sample text.'


# count the words

word_freq = count_words(input_text)


# print the final result

for word, count in word_freq.items():
  print('{word}:{count}')


```

```python
# Exercise 5: check the palindrome

import re

def is_palindrome(text):
  """Checks if a string is a palindrome.

  Args:
    text: The input string.

  Returns:
    True if the string is a palindrome, False otherwise.
  """

  # Preprocessing: Remove non-alphanumeric characters and convert to lowercase
  processed_text = re.sub(r'[^a-zA-Z0-9]', '', text).lower()

  # Compare the processed string with its reverse
  return processed_text == processed_text[::-1]

# Get input from the user
input_string = input("Enter a string: ")

# Check if it's a palindrome and print the result
if is_palindrome(input_string):
  print("True")  # It's a palindrome
else:
  print("False") # It's not a palindrome
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
from scipy.stats import chi2

# Sample data (replace with your actual data)
data = {
    'Device': ['Smart Thermostat', 'Smart Thermostat', 'Smart Light', 'Smart Light', 'Smart Thermostat', 'Smart Light'],
    'Satisfaction': ['High', 'Medium', 'High', 'Low', 'High', 'Medium']
}
df = pd.DataFrame(data)

# 1. State the Hypotheses:
# Null Hypothesis (H0): There is no association between the type of smart home device purchased and customer satisfaction level.
# Alternative Hypothesis (H1): There is an association between the type of smart home device purchased and customer satisfaction level.

# 2. Compute the Chi-Square Statistic:
# Create a contingency table
contingency_table = pd.crosstab(df['Device'], df['Satisfaction'])

# Perform the Chi-Square test
chi2_statistic, p_value, degrees_of_freedom, expected_frequencies = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2_statistic)
print("P-value:", p_value)
print("Degrees of Freedom:", degrees_of_freedom)
print("Expected Frequencies:", expected_frequencies)

# 3. Determine the Critical Value:
# Set the significance level (alpha)
alpha = 0.05

# Calculate the critical value
critical_value = chi2.ppf(1 - alpha, degrees_of_freedom)

print("Critical Value:", critical_value)

# Compare the chi-square statistic to the critical value
if chi2_statistic > critical_value:
    print("Reject the null hypothesis. There is a significant association between device type and satisfaction level.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between device type and satisfaction level.")
```

```python
# conclusion

#Hypothesis Testing: A Chi-square test was conducted to determine if there was a significant association between smart home device type and customer satisfaction level. The results and interpretation were presented
```

Hypothesis Testing

State the Hypothesis Statement:
Null hypothesis(H0) : There is no significant differene between the sample mean and theroetical mean.

Alternative Hypothesis(H1) : There is a significant differnce between sample mean and theroetical mean

```python
import math
import pandas as pd
import numpy as pd
```

```python
# Given data

sample_mean = 3050
theoretical_mean = 1000+5*600
std_dev = 5*25
sample_size = 25
```

```python
# Calculate the t_stastical testing
t_stat = (sample_mean - theoretical_mean)/(std_dev/math.sqrt(sample_size))
t_stat
```

```python
#Null Hypothesis (H0): There is no significant difference between the average weekly operating cost and the theoretical cost (no increase).
#Alternative Hypothesis (H1): The average weekly operating cost is greater than the theoretical cost (there is an increase).
```

```python
# determine the critical value


from scipy import stats

a=0.05
dof = 24 # degree of freddom = sample_size - 1
critical_value = stats.t.ppf(1-a/2,dof)
critical_value
```

```python
if abs(t_stat)>critical_value:
  print("Reject null hypothesis")
else:
  print("accept null hypothesis")
```

Test Statistic (t_stat): This value, calculated as approximately 2.2, represents how far the sample mean is from the theoretical mean in terms of standard error.

Critical Value: This value, which is around 2.064, acts as a threshold. If the absolute value of the test statistic exceeds the critical value, it suggests that the observed difference is unlikely to have occurred due to random chance alone.



Decision Rule

We compare the test statistic with the critical value:

If the test statistic (absolute value) > the critical value, we reject the null hypothesis.
If the test statistic (absolute value) <= the critical value, we fail to reject the null hypothesis.Applying the Decision

Since |t_stat| (2.2) is greater than the critical value (2.064), we should reject the null hypothesis

```python
# Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value
def outlier_detection(data,colname):
  q1 = data[colname].quantile(0.25)
  q3 = data[colname].quantile(0.75)
  iqr = q3 - q1

  upper_extreme = q3 + (1.5 * iqr)
  lower_extreme = q1 - (1.5 * iqr)

  return lower_extreme, upper_extreme,q1,q3
```
