Feature Engineering

creating new features

1. Age group:
Rationale:Grouping age into categories can capture non-linear relationships between age and income. For instance, certain age groups might have higher or lower income tendencies due to career stages or life circumstances.

2.Capital Gain Indicator:

Rationale: A binary indicator for whether an individual has any capital gain might be informative, as those with capital gains could have a higher likelihood of earning more than $50K.

```python
df['age_group']= pd.cut(df['age'],bins=[0,25,45,65,100], labels=['0-25','26-45','46-55','66+'])
```

```python
# Creating capital gain indicator
df['capital_gain_indicator'] = df['capital_gain'].apply(lambda x: 1 if x > 0 else 0)
```

1. Log Transformation of capital-gain:

Rationale: The capital-gain feature is often highly skewed, with most people having zero or low capital gains but a few having very high gains. Applying a log transformation can reduce skewness and make the distribution more normal, which can improve the performance of some machine learning algorithms.

```python
import numpy as np

# Log transformation for capital gain
df['capital_gain'] = df['capital_gain'].apply(lambda x: np.log(x + 1))  # Adding 1 to avoid log(0)
```

```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import IsolationForest
import ppscore as pps


# Assuming 'df' is your preprocessed dataframe

# One-Hot Encode Categorical Features
categorical_features = df.select_dtypes(include=['object', 'category']).columns
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_data = encoder.fit_transform(df[categorical_features])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_features))

# Concatenate Encoded Features with Numerical Features
numerical_features = df.drop(columns=categorical_features.union(['income']))
df_encoded = pd.concat([numerical_features, encoded_df], axis=1)

# Initialize Isolation Forest
iso_forest = IsolationForest(contamination=0.01)  # Adjust contamination as needed

# Fit the model and predict outliers
outliers = iso_forest.fit_predict(df_encoded)

# Remove outliers from the original dataframe
df_cleaned = df[outliers == 1]

print(f"Original dataframe shape: {df.shape}")
print(f"Cleaned dataframe shape: {df_cleaned.shape}")


# Calculate PPS matrix
pps_matrix = pps.matrix(df_cleaned)

# Extract relevant PPS scores (predictive power of each feature on 'income')
pps_income = pps_matrix[pps_matrix['y'] == 'income'][['x', 'ppscore']].sort_values(by='ppscore', ascending=False)

# Calculate correlation matrix
correlation_matrix = df_cleaned.corr()

# Print PPS and correlation for comparison
print("\nPredictive Power Score (PPS) for 'income':\n", pps_income)
print("\nCorrelation Matrix:\n", correlation_matrix)

# You can further analyze and visualize the PPS and correlation matrices
# to gain deeper insights into feature relationships.
```

Predictive Power Score (PPS):

PPS measures the ability of one feature to predict another. It provides a more nuanced view of feature relationships compared to correlation, as it can detect non-linear relationships and is less sensitive to outliers.

Correlation Matrix:

The correlation matrix shows the linear relationship between features. It's a good starting point but might miss complex relationships.
