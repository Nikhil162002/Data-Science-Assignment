```python
x= df.drop(columns='letter')
y = df['letter']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.metrics import classification_report
```

```python
# bulid the ANN model
model = keras.Sequential()
# Add hidden layer
model.add(Dense(units=6,activation='relu'))
# add output layer
model.add(Dense(units=26,activation='softmax'))

#compile the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# fit the model
model.fit(x_train,y_train,epochs=10,batch_size=32,validation_split=0.2)
```

```python
#make prediction on the test data
y_pred = model.predict(x_test)
y_pred
```

```python
y_pred_classes =np.argmax(y_pred,axis=1) # get predicted class labels
y_pred_classes
```

```python
def hyper(hp):
  model = keras.Sequential()

  # Update input_dim to match the actual number of features in x_train
  model.add(Dense(hp.Int('units0',min_value=10,max_value=200,step=10),
                  activation=hp.Choice('activation0',values=['relu','tanh','sigmoid']),
                  input_dim=x_train.shape[1])) # Use x_train.shape[1] to get the correct input dimension
  for i in range(1,hp.Int('num_layers',min_value=1,max_value=10)):
    model.add(Dense(hp.Int('units'+ str(i),min_value=10,max_value=200,step=10),
                    activation=hp.Choice('activation'+ str(i),values=['relu','tanh','sigmoid'])))
  # add output layer
  # Change activation to softmax and units to 26 to match the number of classes
  model.add(Dense(units=26, activation='softmax')) # Update units and activation for multi-class classification

  #compile the model
  model.compile(optimizer=hp.Choice('optimizer',values=['adam','rmsprop','sgd']),
                loss='sparse_categorical_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner = kt.RandomSearch(hyper,objective='val_accuracy',max_trials=10)
```

```python
tuner.search(x_train,y_train,epochs=3,validation_data=(x_test,y_test))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
tuned_model = tuner.get_best_models(num_models=1)[0]
```

```python
tuned_model.fit(x_train,y_train,epochs=50, validation_data=(x_test,y_test))
```

```python
tuned_model.evaluate(x_train,y_train) # Training accuracy
```

```python
y_pred = tuned_model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Get predicted class labels
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Accuracy
accuracy = accuracy_score(y_test, y_pred_classes)
print(f"Accuracy: {accuracy}")

# Precision
precision = precision_score(y_test, y_pred_classes, average='weighted')  # Use weighted average for multi-class
print(f"Precision: {precision}")

# Recall
recall = recall_score(y_test, y_pred_classes, average='weighted')  # Use weighted average for multi-class
print(f"Recall: {recall}")

# F1-score
f1 = f1_score(y_test, y_pred_classes, average='weighted')  # Use weighted average for multi-class
print(f"F1-score: {f1}")
```
